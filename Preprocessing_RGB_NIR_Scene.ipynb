```python
import os
import shutil
import subprocess
import shlex

def createFolderList(path):
    
    #Create array of array of videos.
    foldersList = [];
    
    #List all files in the directory and read points from text files one by one
    for filePath in os.listdir(path):
        #if filePath.endswith(".avi") or filePath.endswith(".mp4"):
        
            fileName = os.path.splitext(filePath)[0]
            # append video found.
            foldersList.append([os.path.join(path,filePath),fileName]);

    return foldersList;


def createImagesList(path, folderName):
    
    #Create array of array of videos.
    imagesList = [];
    
    #List all files in the directory and read points from text files one by one
    for filePath in os.listdir(path):
        #if filePath.endswith(".avi") or filePath.endswith(".mp4"):
        if filePath.endswith(".tiff"):
            fileName = os.path.splitext(filePath)[0]
            # append video found.
            imagesList.append([os.path.join(path,filePath),fileName,folderName]);

    return imagesList;




foldersPath = '/data2/public-dataset/nirscene1'
foldersList = createFolderList(foldersPath)
imageLists = []

for folder in foldersList:
    #print folder[0]
    if folder[1] != 'train' and folder[1] != 'val' and folder[1] != 'test' and \
       folder[1] != 'nir' and folder[1] != 'rgb':
        imageList = createImagesList(folder[0], folder[1])
        imageLists.append(imageList)


    
split_ratio = 0.6 # e.g. train: 0.6, val: 0.2, test: 0.2
#------------------------------------------------------
train_folder = os.path.join(foldersPath,'train')
if not os.path.exists(train_folder):
    os.makedirs(train_folder)

val_folder = os.path.join(foldersPath,'val')  
if not os.path.exists(val_folder):
    os.makedirs(val_folder)
    
test_folder = os.path.join(foldersPath,'test')  
if not os.path.exists(test_folder):
    os.makedirs(test_folder)
    
#------------------------------------------------------    
rgb_folder = os.path.join(foldersPath,'rgb')
if not os.path.exists(rgb_folder):
    os.makedirs(rgb_folder)
    
train_rgb_folder = os.path.join(rgb_folder,'train')
if not os.path.exists(train_rgb_folder):
    os.makedirs(train_rgb_folder)

val_rgb_folder = os.path.join(rgb_folder,'val')  
if not os.path.exists(val_rgb_folder):
    os.makedirs(val_rgb_folder)

test_rgb_folder = os.path.join(rgb_folder,'test')  
if not os.path.exists(test_rgb_folder):
    os.makedirs(test_rgb_folder)

#------------------------------------------------------  
nir_folder = os.path.join(foldersPath,'nir')
if not os.path.exists(nir_folder):
    os.makedirs(nir_folder)

train_nir_folder = os.path.join(nir_folder,'train')
if not os.path.exists(train_nir_folder):
    os.makedirs(train_nir_folder)

val_nir_folder = os.path.join(nir_folder,'val')  
if not os.path.exists(val_nir_folder):
    os.makedirs(val_nir_folder)
    
test_nir_folder = os.path.join(nir_folder,'test')  
if not os.path.exists(test_nir_folder):
    os.makedirs(test_nir_folder)
    
#------------------------------------------------------   


for imageList in imageLists:
    numOfImages = len(imageList)/2
    print numOfImages
    threshold = int(numOfImages*split_ratio)
    if (numOfImages-threshold)%2 == 1:
        threshold += 1
    
    print numOfImages-threshold
    print threshold
    print '---'
    rgb_counter = 1
    nir_counter = 1
    for imageName in imageList:
        folderName = imageName[2]
        image_num = imageName[1].split('_')[0]
        image_label = imageName[1].split('_')[1]
        image_path = os.path.dirname(imageName[0])
        if image_label == 'rgb':
            if rgb_counter <= threshold:
                shutil.copy2( imageName[0], os.path.join( train_rgb_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
                shutil.copy2( os.path.join( image_path, image_num+'_nir.tiff'  ), 
                              os.path.join( train_nir_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
            else:
                if rgb_counter <= threshold+(int(numOfImages-threshold)/2):
                    shutil.copy2( imageName[0], os.path.join( val_rgb_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
                    shutil.copy2( os.path.join( image_path, image_num+'_nir.tiff'  ), 
                                  os.path.join( val_nir_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
                    
                else:
                    shutil.copy2( imageName[0], os.path.join( test_rgb_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
                    shutil.copy2( os.path.join( image_path, image_num+'_nir.tiff'  ), 
                                  os.path.join( test_nir_folder, folderName+'_'+str(rgb_counter)+'.tiff' ))
                    
            rgb_counter += 1
            
#         else:
#             if nir_counter <= threshold:
#                 shutil.copy2( imageName[0], os.path.join( train_nir_folder, folderName+'_'+str(nir_counter)+'.tiff' ))
#             else:
#                 if nir_counter <= threshold+(int(numOfImages-threshold)/2):
#                     shutil.copy2( imageName[0], os.path.join( val_nir_folder, folderName+'_'+str(nir_counter)+'.tiff' ))
                    
#                 else:
#                     shutil.copy2( imageName[0], os.path.join( test_nir_folder, folderName+'_'+str(nir_counter)+'.tiff' ))
  
#             nir_counter += 1
```
